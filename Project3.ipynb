## Convert an Image to Grayscale and Save: 
- Convert a color image to grayscale using OpenCV, save the grayscale image to a new file, and display both images using Matplotlib.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Read the image
image = cv2.imread(r'C:\Users\param\Pictures\Screenshots\Screenshot 2024-10-22 204844.png')

# Convert the image from BGR to RGB color space (OpenCV reads images in BGR by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Convert the image to grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Save the grayscale image
cv2.imwrite('grayscale_image.jpg', image_gray)
```

```python
# Display the images
plt.figure(figsize=(10, 5))

# Display original image
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

# Display grayscale image
plt.subplot(1, 2, 2)
plt.imshow(image_gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.show()
```
